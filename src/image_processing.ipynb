```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
data_path = '../data/'
```

# Read Images from video

```python
vidcap = cv2.VideoCapture(data_path + 'videos/sample_video.mp4')
success, image = vidcap.read()
```

```python
for i in range(50):
    success, image = vidcap.read()
    if not success:
        break
```

```python
plt.imshow(vidcap.read()[1])
```

```python
count = 600
while success and count < 900:
    # Only take every 50th frame
    # for i in range(50):
    #     success, image = vidcap.read()
    #     if not success:
    #         break

    cv2.imwrite(data_path + f'images/seq2/frame{str(count).zfill(3)}.jpg', image)
    success, image = vidcap.read()
    print(f"Read file {count}.")
    count += 1
```

```python
count = 1000
while success and count < 1700:
    # Only take every 50th frame
    for i in range(50):
        success, image = vidcap.read()
        if not success:
            break

    cv2.imwrite(data_path + f'images/frame{count}.jpg', image)
    success, image = vidcap.read()
    print(f"Read file {count}.")
    count += 1
```

# Get normalization numbers

```python
# listdir(data_path + f'images/')
width, height = 1920, 1080
arr = np.zeros((len([x for x in listdir(data_path + f'images/') if x.endswith('.jpg')]), height, width, 3))
for i, filename in enumerate([x for x in listdir(data_path + f'images/') if x.endswith('.jpg')]):
    arr[i] = cv2.imread(data_path + f'images/{filename}') / 255.
```

```python
arr.shape
print(np.std(arr, axis=(0, 1, 2)))
print(np.mean(arr, axis=(0, 1, 2)))
```
